```python
file = pd.read_csv(r"D:\ASHSIH\ENGG\COURSES\MachineLearning\MarksSheet.csv")
file.head()
```

```python
columns = file[['CIE1','CIE2','CIE3','AAT','Theory_Attendance','Lab_Attendance','CIE','SEE','Total']]
columns.head(41)
```

```python
plt.scatter(columns.CIE,columns.Total,  color='blue')
plt.xlabel("CIE")
plt.ylabel("Total")
plt.show()
```

```python
plt.scatter(columns.CIE,columns.SEE,  color='blue')
plt.xlabel("CIE")
plt.ylabel("SEE")
plt.show()
```

```python
msk = np.random.rand(len(file)) < 1.0
train = columns[msk]

file1 = pd.read_csv(r"D:\ASHSIH\ENGG\COURSES\MachineLearning\MarksSheet1.csv")
file1.head(8)
columns1 = file1[['CIE1','CIE2','CIE3','AAT','Theory Attendance','Lab Attendance','CIE','SEE','Total']]
columns1.head(11)
msk1 = np.random.rand(len(file1)) < 1.0
test = columns1[msk1]
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['CIE1','CIE2','CIE3','AAT','Theory Attendance','Lab Attendance','CIE']])
y = np.asanyarray(train[['SEE']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['CIE1','CIE2','CIE3','AAT','Theory Attendance','Lab Attendance','CIE']])
print('predictions:',y_hat)
x = np.asanyarray(test[['CIE1','CIE2','CIE3','AAT','Theory Attendance','Lab Attendance','CIE']])
y = np.asanyarray(test[['SEE']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
